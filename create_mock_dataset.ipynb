```python
import pandas as pd
import random
from faker import Faker
import numpy as np

# Initialize
fake = Faker()
random.seed(42)
np.random.seed(42)

# Options for categorical fields
business_types = ['B2B', 'B2B2C', 'B2C']
industries = ['Software', 'Healthcare', 'Retail', 'Finance', 'Logistics', 'Education']
bbb_ratings = ['A+', 'A', 'B', 'C', 'NR']
titles = ['CEO', 'Founder', 'President', 'Managing Director', 'Co-Founder', 'Owner']

# Create 2000 fake leads
data = []
for _ in range(2000):
    revenue = round(random.uniform(0.05, 10.0), 2)  # in millions
    employees = random.randint(3, 500)
    year_founded = random.randint(1990, 2022)
    business_type = random.choices(business_types, weights=[0.5, 0.3, 0.2])[0]
    industry = random.choice(industries)
    bbb = random.choices(bbb_ratings, weights=[0.3, 0.3, 0.2, 0.1, 0.1])[0]
    title = random.choice(titles)

    row = {
        'Company': fake.company(),
        'Website': fake.url(),
        'Industry': industry,
        'Product/Service Category': f"{industry} Solutions",
        'Business Type': business_type,
        'Employees Count': employees,
        'Revenue': revenue,
        'Year Founded': year_founded,
        'BBB Rating': bbb,
        'Street': fake.street_address(),
        'City': fake.city(),
        'State': fake.state(),
        'Company Phone': fake.phone_number(),
        'Company LinkedIn': f"https://www.linkedin.com/company/{fake.slug()}",
        "Owner's First Name": fake.first_name(),
        "Owner's Last Name": fake.last_name(),
        "Owner's Title": title,
        "Owner's LinkedIn": f"https://www.linkedin.com/in/{fake.slug()}",
        "Owner's Phone Number": fake.phone_number(),
        "Owner's Email": fake.email(),
        "Source": "Mock Generator",
        "Created Date": fake.date_this_year(),
        "Updated": fake.date_this_year(),
        "Actions": ""
    }

    data.append(row)

# Convert to DataFrame and save
df = pd.DataFrame(data)
df.to_csv("mock_leads_dataset_2000.csv", index=False)
```
